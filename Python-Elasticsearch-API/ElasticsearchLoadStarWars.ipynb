# Loads Star Wars data into Elasticsearch
## Notes
### Star Wars API info can be found here : https://swapi.co/
### Code connects to Elasticsearch running on IP 10.0.2.15
### If you are running on localhost the connection call will need to be modified
### There is an error_limit variable that is the number of 404's returned
###  because not all of the ID's retrun results

```python
# imports
from elasticsearch import Elasticsearch
import requests, json
```

```python
#connect to our cluster
from elasticsearch import Elasticsearch
es = Elasticsearch([{'host': '10.0.2.15', 'port': 9200}])
```

### Get the films and put in an index named sw_films

```python
i = 0
# how many 404 errors we will accept before we bail
error_limit = 30
current_errors = 0
#while r.status_code == 200:
while error_limit >= current_errors:
    r = requests.get('http://swapi.co/api/films/'+ str(i))
    print(r.status_code)
    if r.status_code == 200:
        # it's not a 404 so write the data
        es.index(index='sw_films', doc_type='_doc', id=i, body=json.loads(r.content))
        i=i+1
    else:
        # it's not a 200 so don't write anything
        current_errors = current_errors + 1
        i=i+1
    print(i)
```
